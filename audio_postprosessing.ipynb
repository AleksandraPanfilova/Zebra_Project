```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import soundfile as sf
from playsound import playsound
```

```python
playsound('zebra audio sample_Bing_413/all/84002.01.whinnie.trim.wav')
```

```python
# calculate stft
A_whinnie,_ = librosa.load('zebra audio sample_Bing_413/all/84002.01.whinnie.trim.wav')
```

```python
def plot_spectrogram(Y,
                     #sr, 
                     #hop_length, 
                     y_axis="linear"
                    ):
    plt.figure(figsize=(25, 10))
    librosa.display.specshow(Y, 
                             #sr=sr, 
                             #hop_length=hop_length, 
                             x_axis="time",
                             y_axis=y_axis
                            )
    plt.colorbar(format="%+2.f")
```

```python
n = len(A_whinnie)
n_fft = 2048
#A_whinnie_pad = librosa.util.fix_length(A_whinnie, size=n + n_fft // 2)
```

```python
stft_A_whinnie = np.abs(librosa.stft(A_whinnie))
#stft_A_whinnie = np.abs(librosa.stft(A_whinnie_np))
```

```python
stft_A_whinnie = librosa.amplitude_to_db(stft_A_whinnie)
```

```python
plot_spectrogram(stft_A_whinnie)
```

```python
# db to amplitude
stft_A_whinnie_am = librosa.db_to_amplitude(stft_A_whinnie, ref=10)
```

```python
#istft
istft_A_whinnie = librosa.istft(stft_A_whinnie_am)
```

```python
plt.plot(istft_A_whinnie)
```

```python
sf.write('84002.01.whinnie.trim_istft.wav',
        istft_A_whinnie,
        22050)
```
